```python
def multi_plot1():
    import numpy as np
    import matplotlib.pyplot as plt
    from PIL import Image

    img_obj = Image.open(r"img.jpg")
    # img_obj = img_obj.convert("RGB") or "RGBA"
    img = np.array(img_obj)
    img = Image.fromarray(np.uint8(img))

    plt.suptitle("Title for the whole figure")

    ax = plt.subplot(131)  # or plt.subplot(2, 1, 1)  # rows, cols, index
    ax.imshow(img)
    ax.axis("off")
    ax.set_title('Predicted', loc='left')

    ax = plt.subplot(132)  # or plt.subplot(2, 1, 2)  # rows, cols, index
    ax.imshow(img)
    ax.axis("off")
    ax.set_title('Marked', loc='left')

    ax = plt.subplot(133)  # or plt.subplot(2, 1, 2)  # rows, cols, index
    ax.imshow(img)
    ax.axis("off")
    ax.set_title('Real', loc='left')


    plt.tight_layout()
    plt.show()


multi_plot1()
```

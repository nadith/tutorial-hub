```python
def stack1():
    from matplotlib import pyplot as plt
    from matplotlib import style
    
    style.use('ggplot')
    
    days = [1,2,3,4,5]
    
    sleeping = [7,8,6,11,7]
    eating = [2,3,4,3,2]
    working = [7,8,7,2,2]
    playing = [8,5,7,8,13]
        
    # HACK: These empty plots with thick linewidths are for the legend
    # Since there is no way to label a stack plot
    plt.plot([], [], 'm', label='Sleeping', linewidth=5)
    plt.plot([], [], 'c', label='Eating', linewidth=5)
    plt.plot([], [], 'r', label='Working', linewidth=5)
    plt.plot([], [], 'k', label='Playing', linewidth=5)
    
    # Draw the stack plot
    plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])
    
    plt.title('Stack Plot')
    plt.ylabel('Y axis')
    plt.xlabel('X asis')
    
    plt.legend()
    plt.show()

stack1()
```

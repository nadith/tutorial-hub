```python
def pie1():
    from matplotlib import pyplot as plt
    from matplotlib import style
    
    style.use('ggplot')
    
    slices = [7,2,2,13]    
    activities = ['sleeping', 'eating', 'workings', 'playing']
    
    plt.pie(slices, labels=activities, colors=['c','m','r','b'], 
            startangle=90, shadow=True, explode=(0,0.1,0,0), autopct='%1.1f%%')

    plt.title('Pie Plot')    
    plt.legend()
    plt.show()

pie1()
```

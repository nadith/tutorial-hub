```python
def line1():
    from matplotlib import pyplot as plt
    from matplotlib import style
    
    style.use('ggplot')
    
    x = [5, 8, 10]
    y = [5, 8, 10]
    
    x2 = [6, 9, 11]
    y2 = [6, 15, 7]
    
    plt.plot(x, y, 'g', label='line one', linewidth=5)
    plt.plot(x2, y2, 'c', label='line one', linewidth=5)
    
    plt.title('Epic Info')
    plt.ylabel('Y axis')
    plt.xlabel('X asis')
    
    plt.legend()
    plt.grid(True, color='k') # k is black
    plt.show()

line1()
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

# Show Image

# Drawing shapes on the image

```python
img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)

# draw the image
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis

# draw a line
plt.plot([200,300,400],[100,200,300],'c', linewidth=5)

# finally, show the plot
plt.show()

# To show the plot, can also use (not compatible with intellij-jupyter)
# cv2.imshow("masked", img)
# cv2.waitKey(0)

# Save the image
# cv2.imwrite('watchgray.png', img)
```

```python
img = np.zeros((512,512,3), np.uint8)

# draw a diagonal blue line with thickness of 5 px using opencv
cv2.line(img, (0,0), (511,511), (255,0,0), 5)

# draw the image
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')

# draw a line, using plot
plt.plot([200,300,400], [100,200,300], 'c', linewidth=5)

# finally, show the plot
plt.show()
```

```python
height, width = (512,512)
circle = np.zeros( (height,width) )
cv2.circle(circle, (int(width/2), int(height/2)), 280, 255, thickness=-1 )

# draw the image
plt.imshow(circle, cmap = 'gray', interpolation = 'bicubic')

# finally, show the plot
plt.show()
```

# Crop Image

**Cicular Crop**

```python
img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
height, width = img.shape

circle = np.zeros( (height,width), np.uint8 )
cv2.circle(circle, (int(width/2), int(height/2)), 200, 1, thickness=-1 )

masked_img = cv2.bitwise_and(img, img, mask=circle)

# draw the image
plt.imshow(masked_img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis

# finally, show the plot
plt.show()
```
